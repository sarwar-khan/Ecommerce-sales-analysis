```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import plotly.colors as colors
pio.templates.default = "plotly_white"
```

```python
url = "https://github.com/sarwar-khan/Ecommerce-sales-analysis/raw/refs/heads/main/Sample%20-%20Superstore.csv"
data = pd.read_csv(url, encoding="ISO-8859-1")
```

# **Descriptive statistics of the dataset**

```python
data.describe()
```

```python
data.info()
```

** There is no null value in any of the columns**

```python
data.head(3)
```

**The column Order_date in the data set can be use to create new columns like yeara, quarter, month.  which will be very valuable for sales and profit analysis according to time periods.**

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
```

```python
data['Order Month'] = data['Order Date'].dt.month
data['Order Year'] = data['Order Date'].dt.year
data['Order Day'] = data['Order Date'].dt.dayofweek
```

```python
data.head(2)
```

**Now we have three new columns of month, year and day. Using these we can do better analysis **

# **MONTHLY SALES ANALYSIS**

```python
sales_by_month = data.groupby('Order Month')['Sales'].sum().reset_index()
```

```python
sales_by_month
```

```python
fig = px.line(sales_by_month,
              x='Order Month',
              y='Sales',
              title='Monthly Sales Analysis')
fig.show()
```

# **Category wise sales analysis**

```python
category_sales =data.groupby('Category')['Sales'].sum().reset_index()
category_sales
```

```python
figure = px.pie(category_sales,
                names='Category',
                values='Sales',
                hole = 0.4,
                title='Category wise sales analysis')
figure.show()
```

# **Sales analysis based on sub-categories**

```python
sub_category_sales = data.groupby('Sub-Category')['Sales'].sum().reset_index()
sub_category_sales.head()
```

```python
fig=px.bar(sub_category_sales,
           y = 'Sales',
           x = 'Sub-Category',
           title = 'Sales analysis based on sub-categories')
fig.show()
```

# **Monthly profit from sales**

```python
data.head(1)
```

```python
Mothly_profit_sales = data.groupby('Order Month')['Profit'].sum().reset_index()
Mothly_profit_sales.head()
```

```python
fig =px.bar(Mothly_profit_sales,
             x = 'Order Month',
             y = 'Profit',
             title = 'Monthly profit from sales')
fig.show()
```

# **Profits by category**

```python
Profits_category = data.groupby('Category')['Profit'].sum()
Profits_category.head()
```

```python
fig = px.pie(Profits_category,
             values = Profits_category.values,
             names = Profits_category.index,
             hole = 0.4,
             title = 'Profits by category')
fig.show()
```

# **Profits by sub-category**

```python
Profits_by_sub_category = data.groupby('Sub-Category')['Profit'].sum()
Profits_by_sub_category.head()
```

```python
fig = px.bar(Profits_by_sub_category,
             x = Profits_by_sub_category.index,
             y = Profits_by_sub_category.values,
             title = 'Profits by sub-category')
fig.show()
```

# **Sales and profit by customer sgment**

```python
data.head(2)
```

```python
sale_profit_by_segment = data.groupby('Segment')[['Sales','Profit']].sum().reset_index()
sale_profit_by_segment.head()
```

```python
fig = px.bar(
    sale_profit_by_segment,
    x='Segment',
    y=['Sales', 'Profit'],
    barmode='group',  # 'group' places bars side by side
    title="Sales and Profit by Segment",
    labels={'value': 'Amount', 'variable': 'Metric'},
    text_auto=True  # Displays values on bars
)
fig.show()
```

# **Sales to Profit Ratio**

```python
sales_profit_by_segment = data.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
sales_profit_by_segment['Sales_to_Profit_Ratio'] = sales_profit_by_segment['Sales'] / sales_profit_by_segment['Profit']
print(sales_profit_by_segment[['Segment', 'Sales_to_Profit_Ratio']])
```
